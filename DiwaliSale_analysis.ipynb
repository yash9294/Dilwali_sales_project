```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
#drop the unrelated blank column
df.drop(['Status','unnamed1'], axis=1,inplace=True)
```

```python
#check the null value
df.isnull().sum()
```

```python
#fill the amount with some specific data
df=df.copy()
df['Amount']=df['Amount'].fillna(df['Amount'].mean())
```

```python
df.isnull().sum()
#the amount is filled with the avg of the total amount  
```

```python
df.info()
```

```python
#Here we can see that the amount column is a float, so if we want to change the data type of this column, then
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
#if you want to see the name of the columns
df.columns
```

```python
#if we want to change the name of any columns, then
df.rename(columns= {'Marital_Status':'Shadi', 'Zone':'Region'})
```

```python
#The describe() method generates descriptive statistics of DataFrame ( i.e. count, mean, std , ect )
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
ax= sns.countplot(x= 'Gender', data= df)

for bars in ax. containers:
    ax.bar_label(bars)
```

```python
ax= sns.countplot(x= 'Gender', data = df)
```

```python
sale_gen=df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sale_gen )
```

```python
#From the above graphs, we can see that most  of the buyers are females, and even the purchasing power of females is greater than that of men 
```

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
```

```python
ax=sns.countplot(data = df , x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sale_gens= df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sale_gens)
```

```python
#from the above graph, we can see that most of the buyers are aged 26-35
```

```python
#total number of orders from the top 10 states
```

```python
sale_state=df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20 ,5)})
sns.barplot(x='State', y='Orders', data=sale_state)
```

```python
Sale_states=df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=Sale_states, x='State',y='Amount')
```

```python
ax=sns.countplot(data = df , x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
ax=sns.countplot(data=df, x='Occupation')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_labels(bars)
```

```python
sales_state = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='Occupation',y='Amount')
```

```python
sales_Prod = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_Prod, x='Product_Category', y='Amount')
```
